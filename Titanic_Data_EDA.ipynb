# Creating New feature  : Individual_Fare -->

```python
df['Individual_Fare'] = df['Fare'] / (df['SibSp'] + df['Parch'] + 1)
```

```python
df.head()
```

# Creating new feature : Family_Size

```python
df['Family_Size'] = df['SibSp'] + df['Parch'] + 1
```

```python
df.head()
```

# Creating new Feature  : Family_Type
* 1 --> Alone
* 2-4 --> Nuclear_Family
* Greater than 5 --> Large_Family

```python
# Creating Function -->

def Transform_Family_Size(num):
    if num == 1:
        return 'Alone'
    elif num > 1 and num < 5:
        return 'Nuclear'
    else:
        return 'Large'
```

```python
df['Family_Type'] = df['Family_Size'].apply(Transform_Family_Size)
```

```python
df.head()
```

# Bivariate Analysis on Family_Type -->

```python
pd.crosstab(df['Survived'], df['Family_Type'], normalize = 'columns') * 100
```

## *Conclusions :*
* Chances of survival is 57% if you are travelling with a Nuclear family.
